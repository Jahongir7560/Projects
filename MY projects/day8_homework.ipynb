```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins=sns.load_dataset("penguins")
penguins.head()
```

# SEABORN

## A basic scatter plot

```python
sns.scatterplot(x="species", y="bill_length_mm", data=penguins)
plt.show()
```

bu grafikda pingivin turlari bo'yicha uzunliklari ko'rsatilgan, Chinstrap turiga mansublari nisbatan uzunroq ekanligini ko'rishimiz mumkin

## A scatter plot with hue

```python
sns.scatterplot(x="species", y="body_mass_g", data=penguins, hue="sex")
plt.show()
```

bu grafda har bir tur jinsga ajratib olingan va ularning og'irligi grammda berilgan, Gentoo turiga mansub pingivinlar qolgan turlarga nisbatan deyarli ikki barobar og'irroq bo'larkan

## Setting hue order

```python
sns.scatterplot(x="island", y="species", data=penguins, hue="sex", hue_order=["Female","Male"])
plt.show()
```

bu yerda qaysi orollarda qaysi turga mansub pingivinlar yashashi ko'rsatilgan, Torgersen orolida faqat Adelie turidagilar yasharkan

## Specifying hue colors

```python
hue_colors = {"Female": "yellow","Male": "blue"}
sns.scatterplot(x="island",y="body_mass_g", data=penguins, hue="sex",palette=hue_colors)
plt.show()
```

## Using hue with count plots

```python
sns.countplot(x="species", data=penguins, hue="sex")
```

bu grafda turlarga ajratilgan va jins bo'yicha ularning soni qanchaligi berilgan
