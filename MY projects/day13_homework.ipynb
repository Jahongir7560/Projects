# ENTER

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("ecommerce_product_dataset.csv")
df
```

# <span style="color: RED; font-weight: bold; font-size: 40px;">Standardisation</span>

```python
df["NumReviews_Stand"]=(df["NumReviews"]-df["NumReviews"].mean())/df["NumReviews"].std() 
```

```python

fig , ax=plt.subplots(1,2 , figsize=(10,4))
sns.histplot(df["NumReviews"], ax=ax[0],kde=True, color='r')
sns.histplot(df["NumReviews_Stand"],ax=ax[1],kde=True,color="b")
```

# <span style="color: GREEN; font-weight: bold; font-size: 40px;">MinMax Scaling</span>

```python
df["NumReviews_MinMax"]=(df["NumReviews"]-df["NumReviews"].min())/(df["NumReviews"].max()-df["NumReviews"].min())
```

```python
fig , ax=plt.subplots(1,2 , figsize=(10,4))
sns.histplot(df["NumReviews"], ax=ax[0],kde=True, color='r')
sns.histplot(df["NumReviews_MinMax"],ax=ax[1],kde=True,color="y")
```

# <span style="color: MAGENTA; font-weight: bold; font-size: 40px;">Mean Normalisation</span>

```python
df["NumReviews_MeanNorm"]=(df["NumReviews"]-df["NumReviews"].mean())/(df["NumReviews"].max()-df["NumReviews"].min())
```

```python
fig , ax=plt.subplots(1,2 , figsize=(10,4))
sns.histplot(df["NumReviews"], ax=ax[0],kde=True, color='r')
sns.histplot(df["NumReviews_MeanNorm"],ax=ax[1],kde=True,color="y")
```

# <span style="color: blue; font-weight: bold; font-size: 40px;">MaxAbs Scaling</span>

```python
df["NumReviews_MaxAbs"]=(df["NumReviews"]/df["NumReviews"].abs().max())
```

```python
fig , ax=plt.subplots(1,2 , figsize=(10,4))
sns.histplot(df["NumReviews"], ax=ax[0],kde=True, color='r')
sns.histplot(df["NumReviews_MaxAbs"],ax=ax[1],kde=True,color="y")
```

# <span style="color: YELLOW; font-weight: bold; font-size: 40px;">Robust Scaling</span>

```python
df["NumReviews_Robust"]=(df["NumReviews"]-df["NumReviews"].median())/(df["NumReviews"].quantile(0.75)-df["NumReviews"].quantile(0/25))
```

```python
fig , ax=plt.subplots(1,2 , figsize=(10,4))
sns.histplot(df["NumReviews"], ax=ax[0],kde=True, color='r')
sns.histplot(df["NumReviews_Robust"],ax=ax[1],kde=True,color="y")
```

# ALL

```python
fig , ax=plt.subplots(1,5, figsize=(20,4))
sns.histplot(df["NumReviews"], ax=ax[0],kde=True, color='red')
sns.histplot(df["NumReviews_MaxAbs"],ax=ax[1],kde=True,color="yellow")
sns.histplot(df["NumReviews_Stand"], ax=ax[2],kde=True, color='green')
sns.histplot(df["NumReviews_MinMax"], ax=ax[3],kde=True, color='magenta')
sns.histplot(df["NumReviews_MeanNorm"], ax=ax[4],kde=True, color='blue')
```
