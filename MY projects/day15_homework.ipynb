```python
data = pd.read_csv("https://assets.datacamp.com/production/repositories/2210/datasets/794e0759b73a2d80baa5d8fb88636a47965139d3/college_datav3.csv")
```

```python
data['DEBT_MDN'] = pd.to_numeric(data['DEBT_MDN'], errors="coerce")
data['COMPL_RPY_5YR_RT'] = pd.to_numeric(data['COMPL_RPY_5YR_RT'], errors="coerce")
data['MEDIAN_HH_INC'] = pd.to_numeric(data['MEDIAN_HH_INC'], errors="coerce")
# data = data.dropna().reset_index(drop=True)
data
```

```python
data["COMPL_RPY_5YR_RT_NUMERIC"] = pd.to_numeric(data["COMPL_RPY_5YR_RT"], errors='coerce')
data.groupby(by=['REGION', 'Ownership', "Degree_Type"])['COMPL_RPY_5YR_RT_NUMERIC'].agg(['count', 'mean', 'median'])
```

```python
group_cols = ['REGION', 'Ownership', "Degree_Type", "Locale_Short"]
# Fill missing values by group median
data['COMPL_RPY_5YR_RT_NUMERIC_FILLNA'] = (
    data
    .groupby(by=group_cols)['COMPL_RPY_5YR_RT_NUMERIC']
    .transform(lambda x: x.fillna(x.median()))
    )

data['COMPL_RPY_5YR_RT_NUMERIC_FILLNA_MEAN'] = (
    data
    .groupby(by=group_cols)['COMPL_RPY_5YR_RT_NUMERIC']
    .transform(lambda x: x.fillna(x.mean()))
    )
data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))

data['COMPL_RPY_5YR_RT_NUMERIC'].plot(kind='kde', ax=ax, alpha=0.5, color='blue',  label='Numeric')
data['COMPL_RPY_5YR_RT_NUMERIC_FILLNA'].plot(kind='kde', ax=ax, alpha=0.5, color='green',  label='Fillna_Median')
data['COMPL_RPY_5YR_RT_NUMERIC_FILLNA_MEAN'].plot(kind='kde', ax=ax, alpha=0.5, color='red',  label='Fillna_Mean')

ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()
plt.grid()

plt.title('Histograms of NaN, Fillna_Median, Fillna_Mean')
plt.show()
```
