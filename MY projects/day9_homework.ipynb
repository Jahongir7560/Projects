# ENTER

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
glue=sns.load_dataset("glue")
glue.head(10)
```

# SEABORN

## Figure style: "whitegrid

```python
sns.set_style("whitegrid")
sns.catplot(x="Year", y="Score", data=glue, hue="Model")
plt.show()
```

bu grafda yillar kesimida yelimning modellari o'rtasida natijalari ko'rsatilgan

## Other styles

```python
sns.set_style("ticks")
sns.catplot(x="Year", y="Score", data=glue, hue="Model")
plt.show()
```

```python
sns.set_style("dark")
sns.catplot(x="Year", y="Score", data=glue, hue="Model")
plt.show()
```

## Custom palettes

```python
custom_palette = ["red","green","orange","blue","yellow","purple"]
sns.set_palette(custom_palette)
sns.catplot(x="Year", y="Score", data=glue, hue="Model")
plt.show()
```

set palette yordamida ranglar ketma ketligini tanlashimiz mumkin

## Changing the scale

```python
sns.set_context("paper")
sns.catplot(x="Model", y="Score", data=glue, hue="Year")
plt.xticks(rotation=60)
plt.show()
```

## Adding a title to FacetGrid

```python
g=sns.catplot(x="Model", y="Score", data=glue, hue="Year")
g.figure.suptitle("Glue")
plt.xticks(rotation=60)
plt.show()
```

## Adjusting height of title in FacetGrid

```python
g=sns.catplot(x="Model", y="Score", data=glue, hue="Year")
g.figure.suptitle("Glue", y=1.03)
plt.xticks(rotation=60)
plt.show()
```

## Titles for subplots

```python
g=sns.catplot(x="Model", y="Score", data=glue, hue="Year", kind="box", col="Year")
g.figure.suptitle("Glue", y=1.05)
g.set_xticklabels(rotation=60)
plt.show()
```
