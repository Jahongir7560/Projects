## Index larni kichik to'plamlarga ajratish

## ko'p darajali index lar

```python
df.set_index(["work_year", "job_title"])
```

## index bo'yicha sortlash

```python
new=df.set_index("job_title")
new.sort_index()
```

## Slicing the outer index level

```python
new =df.set_index(["work_year", "experience_level"]).sort_index() 
new.loc[2020:2022]
```

```python
sorted_index = df.set_index("salary").sort_index() 
sorted_index
```

## .loc[] + slicing is a power combo


```python
salary_by_jobtitle_workyear = df.pivot_table( 
"salary", index="work_year", columns="job_title") 
salary_by_jobtitle_workyear
```

### .loc[] + slicing is a power combo


```python
salary_by_jobtitle_workyear.loc[2021:2023] 
```

## The axis argument

```python
salary_by_jobtitle_workyear.mean(axis="index")
```
