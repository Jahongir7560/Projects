# ENTER

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seattle_weather=pd.read_csv("seattle_weather.csv")
austin_weather=pd.read_csv("austin_weather.csv")
medals=pd.read_csv("medals_by_country_2016.csv", index_col=0)
histo=pd.read_csv("summer2016.csv", index_col=0)
```

# PLOT

Bu grafikda Austin va Seattle shaharlarinning oylar kesimida harorat o'zgarishi berilgan, Austin shahri Seattle shahriga nisbatan issiqroq ekan

# Bar

## Olympic medals: visualizing the data

```python
fig, ax= plt.subplots()
ax.bar(medals.index, medals["Gold"])
plt.show()
```

bu bar chart yozgi olimpiya o'yinlarida oltin medallar soni davlatlar kesimida berilgan, ko'rishimiz mumkinki Amerika bu borada 136 medal bilan yetakchilik qilmoqda 

## Interlude: rotate the tick labels

```python
fig, ax= plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
plt.show()
```

## GOLD MEDALS BETWEEN MEN AND WOMEN

```python
sex_m = histo["Sex"]=="M"
medal_m = histo["Medal"] == "Gold"
mens_medals = histo[sex_m & medal_m].value_counts()
spot_f = histo["Sex"] == "F"
medal_f = histo["Medal"] == "Gold"
female_medals = histo[spot_f & medal_f].value_counts()

male_count = mens_medals.sum()
female_count = female_medals.sum()

labels = ['Males', 'Females']
counts = [male_count, female_count]

plt.bar(labels, counts, color=['blue', 'yellow'])
plt.xlabel('Sex')
plt.ylabel('Number of Gold Medals')
plt.title('Gold Medals Won by Sex')
plt.show()
```

bu grafikda jins bo'yicha oltin medal olganlar soni taqqoslangan, erkaklar ayollardan 20ta ko'proq oltin medal olgan

## Olympic medals: visualizing all three

```python
fig, ax= plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.bar(medals.index, medals["Bronze"],
bottom=medals["Gold"] + medals["Silver"])
ax.set_xticklabels(medals.index, rotation=90)
plt.show()
```

bu yerda olimpiya o'yinlaridagi jami to'plangan medallar soni berilgan(davlatlar kesimida)

## Adding a legend

```python
fig, ax= plt.subplots()
ax.bar(medals.index, medals["Gold"], label="Gold")
ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"], label="Silver")
ax.bar(medals.index, medals["Bronze"],
bottom=medals["Gold"] + medals["Silver"], label="Bronze")
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
ax.legend()
plt.show()
```

# HISTOGRAM

## A bar chart

```python
sex_m = histo["Sex"]=="M"
spot_m = histo["Sport"] == "Rowing"
mens_rowing = histo[sex_m & spot_m]
spot_m_2 = histo["Sport"] == "Gymnastics"
mens_gymnastics = histo[sex_m & spot_m_2]


fig, ax = plt.subplots()
ax.bar("Rowing",mens_rowing["Height"].mean())
ax.bar("Gymnastics",mens_gymnastics["Height"].mean())
ax.set_ylabel("Height(cm)")
plt.show()

```

bu bar chart da rowing va gymnastics bo'yicha qatnashgan erkaklarning 

## Introducing histograms

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="mens_rowing")
ax.hist(mens_gymnastics["Height"], label="mens_gymnastics")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

## Customizing histograms: setting the number of bins

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="mens_rowing", bins=5)
ax.hist(mens_gymnastics["Height"], label="mens_gymnastics", bins=5)
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

## Histogram with a histtype of step

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="mens_rowing", bins=5, histtype="step")
ax.hist(mens_gymnastics["Height"], label="mens_gymnastics", bins=5, histtype="step")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

## Adding error bars to bar charts

```python
fig, ax = plt.subplots()
ax.bar("Rowing", mens_rowing["Height"].mean(), yerr=mens_rowing["Height"].std())
ax.bar("Gymnastics", mens_gymnastics["Height"].mean(), yerr=mens_gymnastics["Height"].std())
ax.set_ylabel("Height (cm)")
plt.show()
```

## Adding boxplots

```python
fig, ax = plt.subplots()
ax.boxplot([mens_rowing["Height"], mens_gymnastics["Height"]])
ax.set_xticklabels(["Rowing","Gymnastics"]) 
ax.set_ylabel("Height (cm)")
plt.show()
```
